## CHALLENGE 2: A/B Testing Incentives
#### The Data
Challenge 2 builds off of Challenge 1 by introducing a previously withheld set of incentivized consumers.
<br>
<br>
The <b>"challenge2__withheld_consumers.gz"</b> file contains an additional ~10,000 consumers that were withheld from the Challenge 1 dataset. On January 1 2018, these consumers were incentivized with an offer to earn \$5 cash back if they dined at least once that month. Each row represents a single incentivized consumer. Columns:

    - consumer_id: Unique identifier for each consumer
    - gender: encoding of consumer gender as 0 or 1 (labels not important)
    - partner: encoding of the partner (e.g. United Airlines vs Hilton Hotels) to which the consumer belongs (0 to 10, labels not important).
    - join_dt: the date the consumer joined Rewards Network
    - churn_dt: if a consumer formally deletes their account, this is the effective date of that deletion. (We stop tracking their dines after this date, and they recieve no bonuses after this date). Null indicates the consumer is still a member of the program as of Jan 01, 2018 
<br>
The <b>"challenge2__withheld_dines.gz"</b> file contains 10 years of hypothetical daily dining events from the incentivized consumers. This includes both befor, during, and after the incentivized time-frame:

    - consumer_id: Unique identifier for each consumer
    - dine_dt: The date on which the consumer made a purchase at any restaurant on our platform
